```python
import numpy as np
import os.path 

# system sizes in the x, y, z  
nx = 2
ny = 2
nz = 2
# Element size for x, y, z   
cellengthx = list(np.ones(nx)*5e-6)
cellengthy = list(np.ones(ny)*2e-5)
cellengthz = list(np.ones(nz)*1e-7)
#Grid
Grid_data = np.ones((nx,ny,nz))*0

save_path = '/home/links/hm556/DCoding/DGitLab/heat_flow_MK2/bin' # txt file needs to be save ot the bin dir with the thermalflow application

name_of_file = 'Cellengths'

completeName = os.path.join(save_path, name_of_file+".txt")         

file = open(completeName, "w")

#Write cellength to the txt file with each line being a complete array
count = 0
for i in cellengthx:
    if count>0:
        
        file.write(' '+str(i))
    else:
        file.write('\n'+str(i))
        count+=1
count = 0
for i in cellengthy:
    if count>0:
        
        file.write(' '+str(i))
    else:
        file.write('\n'+str(i))
        count+=1
count = 0
for i in cellengthz:
    if count>0:
        
        file.write(' '+str(i))
    else:
        file.write('\n'+str(i))
        count+=1
file.close()


#Writes the Grid material to a txt file
name_of_file = 'GridMaterial'

completeName = os.path.join(save_path, name_of_file+".txt")         

file = open(completeName, "w")


for i in range(0,nx):
    for j in Grid_data[i]:
        for k in j:
            file.write(str(int(k)))
file.close()

'''
#Working on a new format
file = open(completeName, "w")

count = 0
for i in Grid_data:
    count1 = 0
    for j in i:
        count2 = 0
        if count > 0:
            file.write('\n')
            
            for k in j:
                if count2 == 0:
                    file.write(str(k))
                    count2 += 1
                else:
                    file.write(' '+str(k))
            
        else:
            for k in j:
                file.write(str(k))
            count += 1

    file.write('\n')

file.close()
'''
```
