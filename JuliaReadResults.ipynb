```julia
using Plots
File = open(@raw_str "./DTau-6/Df100000/outputs/output_ThinFilmTau-3_freq_100000.00_00") # Open the file

x = range(1, 110, 110) # Create a range for the x-axis
# Read the lines from the file
lines = readlines(File) # break the file into lines

# Close the file
close(File) # Close the file

# Plot each line starting from line 2
for (index, line) in enumerate(lines) # Loop through the lines
    if index % 100000 == 0 # Plot every 100000th line
        # println("Processing line ", index)
        data = split(line, " ") # Split the line by spaces
        # println(data)
        deleteat!(data,findall(x->x=="",data)) # Remove elements that are only spaces
        # println(data)        # println("Processing line ", index)

        data = parse.(Float64, data) # Convert the strings to floats
        # println(data[2:end])
        # Plot the line here
        plot!(x, data[2:end]) # Plot the data starting from the second element
    end
end
display(plot!()) # Display the plot
```
