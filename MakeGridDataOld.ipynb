```python
import numpy as np
import os.path 
```

```python
# Create a 3D grid
nx = 100
ny = 100
nz = 50
lx = 0.1
ly = 0.1
lz = 0.1

grid = np.ones((nx, ny, nz))*4

# Calculate the center of the torus
center_x = nx // 2
center_y = ny // 2
center_z = nz // 2

# Calculate the radius of the torus
torus_radius = 10
tube_radius =5

# Iterate over each point in the grid
for k in range(1,50):
    for j in range(-50,50):
        for i in range(-50,50):
            Radius = 800
            outer = (Radius +50)
            inner = Radius-200
            if ((i**2+j**2) <= outer) and (((i)**2+j**2) >= inner):
                grid[i-50, j-50, k] = 5
```

```python
heat = np.ones((nx,ny,nz))*0
# Iterate over each point in the grid
for k in range(1,50):
    for j in range(-50,50):
        for i in range(-50,50):
            Radius = 800
            outer = (Radius +50)
            inner = Radius-200
            if (((i)**2+j**2) < inner):
                heat[i-50, j-50, k] = 1

```

```python
combined = np.chararray((nx,ny,nz),itemsize=3,)
for k in range(0,nz):
    for j in range(0,ny):
        for i in range(0,nx):
            combined[i,j,k]=(str(int(grid[i,j,k]))+':'+str(int(heat[i,j,k])))
combinedreshaped = combined.decode('utf-8')
print(combinedreshaped)
file = open('system.in', 'w')
file.write(str(nx)+' '+str(ny)+' '+str(nz)+ '\n')
file.write(str(lx)+' '+str(ly)+' '+str(lz)+ '\n')
file.write('\n')
for k in range(0,nz):
    for j in range(0,ny):
        for i in range(0,nx):
            file.write(combinedreshaped[i,j,k])  # Fix: Change parentheses to square brackets
            file.write(' ')
        file.write('\n')
    file.write('\n')
file.close()
```
